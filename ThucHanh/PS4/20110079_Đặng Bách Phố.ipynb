```python
import numpy as np
from PIL import Image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load image and convert to RGB color space
img = Image.open('phongcanh1.jpeg').convert('RGB')
plt.imshow(img)
plt.show()
```

```python

# Convert image data to numpy array
img_data = np.array(img)

# Reshape data to a 2D array of pixels
w, h, d = tuple(img_data.shape)
img_data_2d = np.reshape(img_data, (w*h, d))

# Perform k-means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=0,n_init=10).fit(img_data_2d)

# Replace each pixel value with the value of its assigned cluster center
compressed_img_data = kmeans.cluster_centers_[kmeans.labels_]

# Reshape compressed data to match original image shape
compressed_img_data = np.reshape(compressed_img_data, (w, h, d))

# Convert compressed data back to image format and save
compressed_img = Image.fromarray(np.uint8(compressed_img_data))
compressed_img.save('compressed_image.jpg')
plt.imshow(compressed_img)
plt.show()
```
