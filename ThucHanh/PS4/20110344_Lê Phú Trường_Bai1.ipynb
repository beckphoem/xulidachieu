```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
```

```python
print("Read the Imange")
image = io.imread('OIP.jpg')
io.imshow(image)
io.show()
```

```python
print("Dim of image")
print("shape of image", image.shape)
print("Image is :",type(image))
print("Image is :",image)
rows = image.shape[0]
colums = image.shape[1]
print(" num of rows is",rows)
print(" num of colums is",colums)
```

```python
print("flatten the image ")
image = image.reshape(rows*colums,3)
print("shape of Flatten image is ",image.shape)
print("Flatten image is ",image)
```

```python
print("Implement k-means")
kmeans = KMeans(n_clusters=3)
print(kmeans.fit(image))
```

```python
print("Compress image ")
compressed_image = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = np.clip(compressed_image.astype('uint8'),0,255)
```

```python
print("Reshape the image ")
print("compressed image shape is :", compressed_image.shape)
compressed_image = compressed_image.reshape(rows,colums,3)
```

```python
print("Save and display ")
io.imsave("OIP_compressed_image_3.png",compressed_image)
io.imshow(compressed_image)
io.show()
```
