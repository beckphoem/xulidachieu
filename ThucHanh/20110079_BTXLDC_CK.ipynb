## Bước 1: data set.

## Bước 2: tìm mean
$\bar{x} = \frac{6 -3 - 2 + 7}{4} = 2$

$\bar{y} = \frac{-4 + 5 + 6 -3}{4} = 1$


## Bước 3: Tính Covariance Matrix
ta có các cặp xếp thứ tự :

$(x, x), (x, y), (y, x), (y, y)$

### 3.1) Tính tất cả các Covariance Matrix của các cặp.

ta có công thức tính:
$$cov(x_i, x_j) = \frac{1}{N - 1} \sum^{N}_{k = 1}{(x_{ik}-\bar{x_i})(x_{jk} - \bar{x_j})}$$

ta tính từng cặp cov:

$
\begin{equation} \label{eq1}
\begin{split}
cov(x, x) & = \frac{1}{N - 1} \sum^{N}_{k = 1}{(x_{ik}-\bar{x_i})(x_{jk} - \bar{x_j})} \\
 & = \frac{1}{N - 1} \sum^{N}_{k = 1}{(x_{ik}-\bar{x_i})^2}\\
 & = \frac{1}{4 - 1}{[(6 - 2)^2 + (-3 - 2)^2 + (-2 -2)^2 + (7 - 2)^2 ]} 
 & = \frac{82}{3}\\
\end{split}
\end{equation}$

$\begin{split}
cov(x, y) & = \frac{1}{4 - 1} [(6-2)(-4-1)+(-3-2)(5-1)+(-2-2)(6-1)+(7-2)(-3-1)] \\
& = -\frac{80}{3}
\end{split}$

$\begin{split}
cov(y, x) & = cov(x, y) = - \frac{80}{3}
\end{split}$

$
\begin{split}
cov(y, y) & = \frac{1}{N - 1} \sum^{N}_{k = 1}{(y_{ik}-\bar{y_i})^2}\\
& = \frac{1}{4 - 1}[(-4-1)^2 + (5-1)^2 + (6-1)^2 + (-3-1)^2] = \frac{82}{3}\\
\end{split}$

### 3.2) Covariance Matrix
$$ 
\begin{split}
C & = \begin{bmatrix} 
cov(x, x) & cov(x, y) \\
cov(y, x) & cov(y,y)\end{bmatrix} \\
& = \begin{bmatrix} 
\frac{82}{3} & -\frac{80}{3} \\
-\frac{80}{3} & \frac{82}{3}\end{bmatrix}\\
\end{split}$$

 ## Bước 4: Eigen value, Eigen vector
 ### 4.1) Eigen value
 
 $$\begin{align} 
 & \det(C -\lambda I) &= 0 \\
 \Leftrightarrow & \det\left(C - \lambda\begin{bmatrix}1 & 0 \\0 & 1 \end{bmatrix}\right) &= 0\\
\Leftrightarrow & \det\left(C - \begin{bmatrix}\lambda & 0 \\0 & \lambda \end{bmatrix}\right) &= 0 \\
\Leftrightarrow & \det\left(\begin{bmatrix} \frac{82}{3} & -\frac{80}{3} \\
-\frac{80}{3} & \frac{82}{3}\end{bmatrix} - \begin{bmatrix}\lambda & 0 \\0 & \lambda \end{bmatrix}\right) &= 0 \\
\Leftrightarrow & \det\left(\begin{bmatrix} 
\frac{82}{3}-\lambda & -\frac{80}{3} \\
-\frac{80}{3} & \frac{82}{3}-\lambda\\ \end{bmatrix} \right) &= 0 \\
\Leftrightarrow & (\frac{82}{3}-\lambda)(\frac{82}{3}-\lambda) - (-\frac{80}{3} )(-\frac{80}{3}) &= 0 \\
\Leftrightarrow & (\frac{82}{3}-\lambda)^2 - (-\frac{80}{3} )^2 &= 0 \\
\Leftrightarrow & (\frac{82}{3})^2-2\frac{82}{3}\lambda + \lambda^2 - (-\frac{80}{3} )^2 &= 0 \\
\Leftrightarrow & \frac{6724}{9}-\frac{164}{3}\lambda + \lambda^2 -\frac{6400}{9} &= 0 \\
\Leftrightarrow & \lambda^2 -\frac{164}{3}\lambda + 36 &= 0 \\
\Leftrightarrow & \lambda_1 = 54, \lambda_2 =\frac{2}{3}, \lambda_1 > \lambda_2 \Rightarrow \lambda_1
 \end{align}$$
 
 
 
 

### 4.2) Eigen vector of $\lambda_1$
$$\begin{align}

\Leftrightarrow& \begin{bmatrix}

\begin{bmatrix} u_1 \\ u_2 \end{bmatrix} &= 0\\

\end{align}$$

$$\begin{align}
(\frac{82}{3}-\lambda_1)u_1 - (-\frac{80}{3})u_2 & = 0 \\
(-\frac{80}{3})u_1 - (\frac{82}{3}-\lambda_1)u_2 & = 0
\end{align} $$

$\begin{align} 
(\frac{82}{3}-\lambda_1)u_1 - (-\frac{80}{3})u_2 & = 0 \\
(\frac{82}{3}-\lambda_1)u_1 &=(-\frac{80}{3})u_2 \\
\frac{u_1}{-\frac{80}{3}} &= \frac{u_2}{(\frac{82}{3}-\lambda_1)} = t
\end{align}$

## 2.3 c) Hãy so sánh và đưa ra nhận xét the performance giữa PCA và ICA.

PCA (Principal Component Analysis) và ICA (Independent Component Analysis) là hai phương pháp phân tích thành phần được sử dụng để giảm chiều dữ liệu và tìm ra các thành phần quan trọng. Dưới đây là một số điểm để so sánh và đưa ra nhận xét về hiệu suất của PCA và ICA:

1. Mục đích chính:
   - PCA: Mục đích chính của PCA là tìm ra các thành phần chính (principal components) mà giữ lại phần lớn thông tin biến thiên trong dữ liệu. PCA tập trung vào việc tối đa hóa phương sai (variance) giữa các thành phần.
   - ICA: Mục đích chính của ICA là tìm ra các thành phần độc lập (independent components), giả sử rằng các tín hiệu quan sát được là sự pha trộn tuyến tính của các nguồn độc lập. ICA tập trung vào việc phân tách và tìm ra các thành phần độc lập trong dữ liệu.

2. Độc lập:
   - PCA: Các thành phần chính trong PCA không nhất thiết phải độc lập với nhau. Các thành phần chính được sắp xếp theo độ giảm dần của phương sai.
   - ICA: Các thành phần độc lập trong ICA độc lập với nhau và không phụ thuộc tuyến tính vào nhau.

3. Giả định:
   - PCA: PCA giả định rằng các thành phần chính là tuyến tính và phân phối theo Gaussian.
   - ICA: ICA giả định rằng các tín hiệu quan sát được là sự pha trộn tuyến tính của các tín hiệu nguồn độc lập.

4. Ứng dụng:
   - PCA: PCA thường được sử dụng cho việc giảm chiều dữ liệu, trực quan hóa dữ liệu, và phân tích đặc trưng trong các bài toán như phân tích thành phần chính trong thị giác máy tính, xử lý ảnh, và nhận dạng giọng nói.
   - ICA: ICA thường được sử dụng trong xử lý tín hiệu, trích xuất nguồn và phân tích dữ liệu nơ-ron, cũng như trong các lĩnh vực như phân tích tín hiệu hình ảnh và nhiễu, phân tích tín hiệu âm thanh và nhiễu, và nhận dạng mẫu.

5. Hiệu suất:
   - PCA: PCA giảm chiều dữ liệu bằng cách chọn các thành phần chính với

 phương sai lớn nhất. Nó có thể giúp giữ lại thông tin quan trọng nhưng có thể làm mất đi sự đặc trưng của dữ liệu trong quá trình giảm chiều.
   - ICA: ICA tìm ra các thành phần độc lập trong dữ liệu, tách riêng các tín hiệu nguồn ban đầu. Nó có thể hữu ích trong việc phân tách và trích xuất các tín hiệu nguồn, nhưng cũng có thể bị ảnh hưởng bởi các yếu tố như mất mát thông tin hoặc nhiễu.

Tổng quan, PCA và ICA đều là các phương pháp quan trọng trong phân tích thành phần và giảm chiều dữ liệu. PCA tập trung vào việc tối ưu hóa phương sai giữa các thành phần, trong khi ICA tập trung vào việc phân tách và tìm ra các thành phần độc lập.

## 3.2 b) Từ dữ liệu được trích xuất ở câu trên, hãy dùng k-Means clustering để phân thành 3 cụm và vẽ hình biểu diễn.

```python
from sklearn.cluster import KMeans
```

```python
# Tạo ma trận dữ liệu
dt = np.column_stack((ApplicantIncome, LoanAmount))

# Áp dụng k-Means clustering với số cụm là 3
kmeans = KMeans(n_clusters=3)
kmeans.fit(dt)

# Lấy thông tin về các trung tâm cụm và nhãn của các điểm dữ liệu
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Vẽ scatter plot với các cụm đã được phân loại màu khác nhau
plt.scatter(dt[:, 0], dt[:, 1], c=labels)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', color='red', label='Cluster Centers')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.title('k-Means Clustering')

# Hiển thị biểu đồ
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1000, min_samples=2)
dbscan.fit(dt)

# Lấy thông tin về nhãn của các điểm dữ liệu
labels = dbscan.labels_

# Số cụm (bao gồm cụm nhiễu)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print("Number of clusters:", n_clusters)

# Vẽ scatter plot với các cụm đã được phân loại màu khác nhau
plt.scatter(dt[:, 0], dt[:, 1], c=labels)
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.title('DBSCAN Clustering')

# Hiển thị biểu đồ
plt.show()
```

1. Nguyên tắc phân cụm:

   - DBSCAN: DBSCAN có khả năng xử lý cụm có hình dạng và kích thước không đều. Nó cũng có thể xử lý các cụm có hình dạng phức tạp và không cần phải giả định về kích thước và hình dạng của các cụm.

 tham số này có thể ảnh hưởng đáng kể đến kết quả phân cụm.

   - k-Means: k-Means không có khả năng xử lý nhiễu một cách tự động. Nếu có dữ liệu nhiễu, nó có thể ảnh hưởng đến kết quả phân cụm.
   - DBSCAN: DBSCAN có khả năng xác định và loại bỏ dữ liệu nhiễu một cách tự động. Nó phân loại các điểm dữ liệu mà không thuộc vào bất kỳ cụm nào là nhiễu.

Tổng quan, k-Means và DBSCAN có cách tiếp cận và tính chất khác nhau. k-Means là phương pháp phân cụm tuyến tính với số lượng cụm xác định trước, trong khi DBSCAN là phương pháp phân cụm dựa trên mật độ dữ liệu và có khả năng xác định số lượng cụm tự động và xử lý nhiễu.
```python
import numpy as np
from sklearn.decomposition import NMF
from skimage.io import imread, imsave

def compress_image(image_path, n_components):
    # Đọc ảnh và chuyển đổi thành ma trận numpy
    image = imread(image_path)
    image = image.astype(float)

    # Chia ma trận ảnh thành các khối con kích thước 8x8
    block_size = 8
    height, width, _ = image.shape
    n_blocks_h = height // block_size
    n_blocks_w = width // block_size
    image_blocks = image[:n_blocks_h*block_size, :n_blocks_w*block_size, :].reshape(n_blocks_h, block_size, n_blocks_w, block_size, 3)
    image_blocks = image_blocks.transpose(0, 2, 1, 3, 4).reshape(-1, block_size, block_size, 3)

    # Chuẩn bị ma trận dữ liệu đầu vào cho NMF
    X = image_blocks.reshape(-1, block_size * block_size * 3)

    # Áp dụng NMF
    model = NMF(n_components=n_components, init='random', random_state=0)
    W = model.fit_transform(X)
    H = model.components_

    # Tái tạo ảnh từ ma trận cơ sở và ma trận hệ số
    X_reconstructed = np.dot(W, H)
    X_reconstructed = np.clip(X_reconstructed, 0, 255)
    X_reconstructed = X_reconstructed.reshape(-1, block_size, block_size, 3)
    X_reconstructed = X_reconstructed.transpose(0, 2, 1, 3).reshape(n_blocks_h, n_blocks_w, block_size, block_size, 3)
    reconstructed_image = X_reconstructed.transpose(0, 2, 1, 3, 4).reshape(height, width, 3)

    # Lưu ảnh tái tạo
    imsave('reconstructed_image.jpg', reconstructed_image.astype(np.uint8))

# Đường dẫn đến ảnh gốc
image_path = image_link

# Số thành phần trong NMF (số lượng đặc trưng)
n_components = 50

# Thực hiện nén ảnh
compress_image(image_path, n_components)
```

## 4.2 b) Hãy nén ảnh trên bằng việc sử dụng truncated singular value decomposition.

```python
import numpy as np
from skimage.io import imread, imsave

def compress_image(image_path, k):
    # Đọc ảnh và chuyển đổi thành ma trận numpy
    image = imread(image_path)
    image = image.astype(float)

    # Chuyển đổi ảnh màu sang grayscale nếu cần thiết
    if len(image.shape) > 2:
        image = np.mean(image, axis=2)

    # Áp dụng Singular Value Decomposition (SVD)
    U, S, V = np.linalg.svd(image)

    # Truncate SVD bằng cách giữ lại k giá trị riêng hàng đầu
    U_truncated = U[:, :k]
    S_truncated = np.diag(S[:k])
    V_truncated = V[:k, :]
    
    # Tái tạo ảnh từ phần trùng lặp đã được giữ lại
    reconstructed_image = U_truncated @ S_truncated @ V_truncated

    # Chuẩn hóa giá trị pixel trong khoảng từ 0 đến 255
    reconstructed_image = np.clip(reconstructed_image, 0, 255)
    reconstructed_image = reconstructed_image.astype(np.uint8)

    # Lưu ảnh tái tạo
    imsave('reconstructed_image2.jpg', reconstructed_image)

# Đường dẫn đến ảnh gốc
image_path = image_link

# Số lượng giá trị riêng giữ lại (thấp hơn làm giảm kích thước ảnh nén)
k = 50

# Thực hiện nén ảnh
compress_image(image_path, k)
```

```python
from skimage.io import imread, imshow
reconstructed_image_path = 'reconstructed_image2.jpg'
reconstructed_image = imread(reconstructed_image_path)
imshow(reconstructed_image)
```

## 4.3c) So sánh the performance giữa NMF và truncated singular value decomposition.
Phương pháp NMF (Non-negative Matrix Factorization) và tSVD (truncated Singular Value Decomposition) đều là các phương pháp phân tích ma trận được sử dụng trong việc nén ảnh. Dưới đây là một số điểm so sánh về hiệu suất giữa hai phương pháp này:

1. Hiệu quả nén: Trong nhiều trường hợp, tSVD có thể đạt được hiệu quả nén tốt hơn so với NMF. Điều này do tSVD giảm số lượng giá trị riêng và sử dụng ít thông tin hơn để tái tạo ảnh. Trong khi đó, NMF tạo ra một phân rã không âm của ma trận và có thể yêu cầu số lượng thành phần lớn hơn để đạt được chất lượng tái tạo tương tự.

2. Thời gian chạy: NMF thường mất nhiều thời gian để tính toán hơn so với tSVD. Điều này do quá trình tối ưu hóa không âm của NMF yêu cầu các vòng lặp để đạt được kết quả tốt nhất. Trong khi đó, tSVD chỉ thực hiện phân rã ma trận SVD và trích xuất một số thành phần riêng hàng đầu.

3. Khả năng tái tạo ảnh: tSVD có xu hướng giữ lại các thông tin quan trọng nhất trong ảnh, nhưng có thể mất mát một số chi tiết nhỏ. NMF có khả năng giữ lại các cấu trúc phi tuyến và chi tiết nhỏ hơn, nhưng cũng có thể tạo ra nhiễu trong quá trình tái tạo.

4. Ứng dụng: cả NMF và tSVD được sử dụng trong nén ảnh và xử lý tín hiệu, nhưng có các ứng dụng khác nhau. NMF thường được sử dụng trong việc tìm kiếm đặc trưng và phân tích nội tại dữ liệu, trong khi tSVD thường được sử dụng trong việc nén và khai phá dữ liệu.

Tuy nhiên, hiệu suất của NMF và tSVD cũng phụ thuộc vào cách cài đặt và các tham số được chọn. Điều này có nghĩa là kết quả có thể thay đổi tùy thuộc vào bài toán cụ thể và tập dữ liệu đang được xử lý. Do đó, việc thử nghiệm và tinh chỉnh các tham số là quan trọng để đạt được kết quả tốt nhất cho từng phương pháp.
